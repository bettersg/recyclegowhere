```python
def get_nodes_involved(way = None, relation = None):
    if way is not None:
        return way['nodeRefs']
    else:
        ls = []
        for w in relation['members']:
            if w['type'] == 'way':
                for item in ways.loc[w['ref']]['nodeRefs']:
                    ls.append(item)
            else:
                ls.append(w['ref'])
        return ls
    
def nodeID_to_coords(nodeIDs: list):
    coords_ls_x = []
    coords_ls_y = []
    for i in nodeIDs:
        try:
            coords = nodes.loc[i]['lat'], nodes.loc[i]['lon']
            coords_ls_x.append(coords[0])
            coords_ls_y.append(coords[1])
        except:
            continue
    return coords_ls_x, coords_ls_y
    

    
```

```python
def get_routes(relation, kind = 'bus'):
    ls2 = []
    for i,r in relation.iterrows():
        if 'route' in r['tags'].keys():
            if 'network' in r['tags'].keys():
                if r['tags']['route'] == kind and r['tags']['network'] in ['Singapore Rail',
                                                                           'Land Transport Authority',
                                                                          'Circle Line (CCL)/Downtown Line (DTL)',
                                                                           'Downtown Line (DTL)',
                                                                           'East West Line (EWL)/North East Line (NEL)',
                                                                          'MRT',
                                                                           'North South Line (NSL)/Circle Line (CCL)',
                                                                           'Singapore',
                                                                           'SBS']:    
                    ls = []
                    for member in r['members']:
                        if member['type'] == 'node' and member['role'] != 'stop_inactive':
                            ls.append(member['ref'])
                    ls2.append(ls)
    return ls2

#get the operating routes only
bus_routes = get_routes(relations, 'bus')
train_routes = get_routes(relations, 'subway')
#rail_routes = get_routes(relations, 'railway')
monorail_routes = get_routes(relations, 'monorail')

print(len(bus_routes)) #723 different routes
print(len(train_routes)) #18 different routes
print(len(monorail_routes)) #10 monorail routes
```
